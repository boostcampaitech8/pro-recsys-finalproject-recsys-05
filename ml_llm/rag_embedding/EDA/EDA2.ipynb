```python
# =============================================================================
# 📦 필요한 라이브러리 import
# =============================================================================
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, defaultdict
from datetime import datetime
from itertools import combinations
import re
import warnings
warnings.filterwarnings('ignore')

# 폰트 설정 (한글 사용 안함)
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)

# =============================================================================
# 📂 데이터 로드
# =============================================================================
def load_jsonl(filepath):
    """JSONL 파일을 DataFrame으로 로드"""
    data = []
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return pd.DataFrame(data)

print("="*80)
print("🎮 STEAM GAMES 데이터셋 로딩 중...")
print("="*80)

# 데이터 로드
df = load_jsonl('/data/ephemeral/home/T8161/game_data/extracted/steam_games_info.jsonl')

print(f"\n✅ 데이터 로드 완료!")
print(f"   총 게임 수: {len(df):,}개")
print(f"   총 컬럼 수: {len(df.columns)}개")

# =============================================================================
# 🔧 전처리
# =============================================================================
def extract_year(date_str):
    """출시일에서 연도 추출"""
    if pd.isna(date_str) or not isinstance(date_str, str):
        return None
    try:
        for fmt in ['%Y-%m-%d', '%d %b, %Y', '%b %d, %Y', '%Y']:
            try:
                return datetime.strptime(date_str, fmt).year
            except:
                continue
        if date_str.isdigit() and len(date_str) == 4:
            return int(date_str)
        return None
    except:
        return None

# age_rating을 숫자로 변환
def parse_age_rating(age):
    if pd.isna(age):
        return 0
    if isinstance(age, (int, float)):
        return int(age)
    if isinstance(age, str):
        numbers = re.findall(r'\d+', str(age))
        if numbers:
            return int(numbers[0])
    return 0

# 전처리 실행
df['release_year'] = df['release_date'].apply(extract_year)
df['age_rating_num'] = df['age_rating'].apply(parse_age_rating)
df['desc_en_length'] = df['short_description_en'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)
df['desc_kr_length'] = df['short_description_kr'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)

print(f"\n✅ 전처리 완료!")

# =============================================================================
# 📊 전체 분석 시작
# =============================================================================
print("\n" + "="*80)
print("📊 STEAM GAMES 데이터셋 전체 분석 리포트")
print("="*80)
print(f"분석 일시: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"데이터 파일: steam_games_info.jsonl")
print("="*80)

# =============================================================================
# SECTION 1: 🆔 기본 정보 & 식별
# =============================================================================
print("\n\n" + "="*80)
print("SECTION 1: 기본 정보 & 식별")
print("="*80)

# 1-1. 게임 타입 (type)
print("\n[1-1] 컨텐츠 타입 분포")
print("-" * 80)
type_dist = df['type'].value_counts()
print(f"\n총 {len(type_dist)}개 타입:")
for game_type, count in type_dist.items():
    pct = (count / len(df)) * 100
    print(f"  {game_type:20s}: {count:7,}개 ({pct:6.2f}%)")

# 그래프: 게임 타입 분포 (영어로)
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
type_dist_top = type_dist.head(7)
colors = plt.cm.Set3(range(len(type_dist_top)))
ax.barh(range(len(type_dist_top)), type_dist_top.values, color=colors)
ax.set_yticks(range(len(type_dist_top)))
ax.set_yticklabels(type_dist_top.index)
ax.set_xlabel('Number of Games')
ax.set_title('Game Type Distribution', fontsize=14, fontweight='bold')
ax.invert_yaxis()
for i, v in enumerate(type_dist_top.values):
    ax.text(v, i, f' {v:,} ({v/len(df)*100:.1f}%)', va='center')
plt.tight_layout()
plt.show()

# 1-2. 한국 스토어 접근 가능 여부
print("\n[1-2] 한국 스토어 접근 가능 여부 (is_available_in_kr)")
print("-" * 80)
kr_available = df['is_available_in_kr'].value_counts()
for status, count in kr_available.items():
    pct = (count / len(df)) * 100
    status_label = "접근 가능" if status else "접근 불가"
    print(f"  {status_label:15s}: {count:7,}개 ({pct:6.2f}%)")

# 1-3. 출시일 분석
print("\n[1-3] 출시일 (release_date) 분석")
print("-" * 80)

year_valid = df['release_year'].notna().sum()
year_missing = df['release_year'].isna().sum()

print(f"\n출시 연도 정보:")
print(f"  데이터 있음: {year_valid:,}개 ({year_valid/len(df)*100:.2f}%)")
print(f"  데이터 없음: {year_missing:,}개 ({year_missing/len(df)*100:.2f}%)")

if year_valid > 0:
    print(f"\n연도 범위:")
    print(f"  최초: {int(df['release_year'].min())}년")
    print(f"  최근: {int(df['release_year'].max())}년")
    
    # 연도별 게임 수
    year_counts = df['release_year'].value_counts().sort_index()
    
    print(f"\n📊 연도별 출시 게임 수 Top 10:")
    for year, count in year_counts.nlargest(10).items():
        pct = (count / len(df)) * 100
        print(f"  {int(year)}년: {count:5,}개 ({pct:5.2f}%)")
    
    # 연도대별 분포
    print(f"\n📊 연도대별 분포:")
    decade_ranges = [
        (1990, 1999, "1990s"),
        (2000, 2009, "2000s"),
        (2010, 2014, "2010-2014"),
        (2015, 2019, "2015-2019"),
        (2020, 2024, "2020-2024"),
        (2025, 2030, "2025+")
    ]
    
    decade_data = []
    for start, end, label in decade_ranges:
        count = ((df['release_year'] >= start) & (df['release_year'] <= end)).sum()
        pct = (count / year_valid) * 100 if year_valid > 0 else 0
        print(f"  {label:15s}: {count:6,}개 ({pct:5.2f}%)")
        decade_data.append((label, count))
    
    # 그래프 1: 연도별 게임 출시 추이 (영어로)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # 왼쪽: 연도별 추이
    year_counts_plot = year_counts[year_counts.index >= 1997]
    ax1.plot(year_counts_plot.index, year_counts_plot.values, marker='o', linewidth=2, markersize=4)
    ax1.fill_between(year_counts_plot.index, year_counts_plot.values, alpha=0.3)
    ax1.set_xlabel('Release Year', fontsize=11)
    ax1.set_ylabel('Number of Games', fontsize=11)
    ax1.set_title('Game Release Trend (1997-2026)', fontsize=14, fontweight='bold')
    ax1.grid(True, alpha=0.3)
    ax1.tick_params(axis='x', rotation=45)
    
    # 오른쪽: 연도대별 분포
    labels, counts = zip(*decade_data)
    colors = plt.cm.Spectral(np.linspace(0, 1, len(labels)))
    ax2.barh(range(len(labels)), counts, color=colors)
    ax2.set_yticks(range(len(labels)))
    ax2.set_yticklabels(labels)
    ax2.set_xlabel('Number of Games', fontsize=11)
    ax2.set_title('Games by Decade', fontsize=14, fontweight='bold')
    ax2.invert_yaxis()
    for i, v in enumerate(counts):
        ax2.text(v, i, f' {v:,}', va='center')
    
    plt.tight_layout()
    plt.show()

# 1-4. 연령 등급 (age_rating)
print("\n[1-4] 연령 등급 (age_rating) 분석")
print("-" * 80)

age_dist = df['age_rating_num'].value_counts().sort_index()
print(f"\n총 {len(age_dist)}개 등급:")

age_data_for_print = []
age_labels_en = []  # 영어 레이블
for age in sorted(age_dist.index):
    count = age_dist[age]
    pct = (count / len(df)) * 100
    
    if age == 0:
        label = "전체이용가/정보없음"
        label_en = "All Ages/No Info"
    else:
        label = f"{age}세 이용가"
        label_en = f"Age {age}+"
    
    if pct >= 0.05:  # 0.05% 이상만 출력
        print(f"  {label:20s}: {count:7,}개 ({pct:6.2f}%)")
        age_data_for_print.append((label_en, count))  # 영어 레이블 저장
        age_labels_en.append(label_en)

adult_count = (df['age_rating_num'] >= 18).sum()
adult_pct = (adult_count / len(df)) * 100
print(f"\n  → 성인용(18세+): {adult_count:,}개 ({adult_pct:.2f}%)")

# 그래프: 연령 등급 분포 (영어로)
fig, ax = plt.subplots(figsize=(10, 6))
labels, counts = zip(*age_data_for_print[:10])  # Top 10만
colors = plt.cm.RdYlGn_r(np.linspace(0.2, 0.8, len(labels)))
ax.bar(range(len(labels)), counts, color=colors, edgecolor='black', linewidth=0.5)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_ylabel('Number of Games', fontsize=11)
ax.set_title('Age Rating Distribution', fontsize=14, fontweight='bold')
ax.grid(True, axis='y', alpha=0.3)
for i, v in enumerate(counts):
    ax.text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()

# 1-5. 개발사 (developers) 분석
print("\n[1-5] 개발사 (developers) 분석")
print("-" * 80)

all_developers = []
for devs in df['developers']:
    if isinstance(devs, list) and len(devs) > 0:
        all_developers.extend(devs)

dev_counts = Counter(all_developers)
print(f"\n총 고유 개발사 수: {len(dev_counts):,}개")
print(f"개발사 정보 있는 게임: {(df['developers'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}개")

print(f"\n📊 Top 20 개발사 (게임 수 기준):")
for i, (dev, count) in enumerate(dev_counts.most_common(20), 1):
    pct = (count / len(df)) * 100
    print(f"  {i:2d}. {dev[:50]:50s}: {count:4,}개 ({pct:5.2f}%)")

# 그래프: Top 15 개발사 (영어로)
fig, ax = plt.subplots(figsize=(12, 8))
top_devs = dev_counts.most_common(15)
devs, counts = zip(*top_devs)
devs_short = [d[:40] + '...' if len(d) > 40 else d for d in devs]
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(devs)))
ax.barh(range(len(devs_short)), counts, color=colors)
ax.set_yticks(range(len(devs_short)))
ax.set_yticklabels(devs_short)
ax.set_xlabel('Number of Games', fontsize=11)
ax.set_title('Top 15 Developers (by Game Count)', fontsize=14, fontweight='bold')
ax.invert_yaxis()
for i, v in enumerate(counts):
    ax.text(v, i, f' {v:,}', va='center')
plt.tight_layout()
plt.show()

# 1-6. 퍼블리셔 (publishers) 분석
print("\n[1-6] 퍼블리셔 (publishers) 분석")
print("-" * 80)

all_publishers = []
for pubs in df['publishers']:
    if isinstance(pubs, list) and len(pubs) > 0:
        all_publishers.extend(pubs)

pub_counts = Counter(all_publishers)
print(f"\n총 고유 퍼블리셔 수: {len(pub_counts):,}개")
print(f"퍼블리셔 정보 있는 게임: {(df['publishers'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}개")

print(f"\n📊 Top 20 퍼블리셔 (게임 수 기준):")
for i, (pub, count) in enumerate(pub_counts.most_common(20), 1):
    pct = (count / len(df)) * 100
    print(f"  {i:2d}. {pub[:50]:50s}: {count:4,}개 ({pct:5.2f}%)")

# 그래프: Top 15 퍼블리셔 (영어로)
fig, ax = plt.subplots(figsize=(12, 8))
top_pubs = pub_counts.most_common(15)
pubs, counts = zip(*top_pubs)
pubs_short = [p[:40] + '...' if len(p) > 40 else p for p in pubs]
colors = plt.cm.plasma(np.linspace(0.3, 0.9, len(pubs)))
ax.barh(range(len(pubs_short)), counts, color=colors)
ax.set_yticks(range(len(pubs_short)))
ax.set_yticklabels(pubs_short)
ax.set_xlabel('Number of Games', fontsize=11)
ax.set_title('Top 15 Publishers (by Game Count)', fontsize=14, fontweight='bold')
ax.invert_yaxis()
for i, v in enumerate(counts):
    ax.text(v, i, f' {v:,}', va='center')
plt.tight_layout()
plt.show()

# 1-7. DLC 개수 (dlc_count) 분석
print("\n[1-7] DLC 개수 (dlc_count) 분석")
print("-" * 80)

dlc_stats = df['dlc_count']
print(f"\nDLC 통계:")
print(f"  평균 DLC 수: {dlc_stats.mean():.2f}개")
print(f"  중앙값: {dlc_stats.median():.0f}개")
print(f"  최대: {dlc_stats.max()}개")

games_with_dlc = df[df['dlc_count'] > 0]
print(f"\nDLC 보유 게임: {len(games_with_dlc):,}개 ({len(games_with_dlc)/len(df)*100:.2f}%)")

print(f"\n📊 DLC 개수 분포:")
dlc_ranges = [
    (0, 0, "No DLC"),
    (1, 5, "1-5"),
    (6, 10, "6-10"),
    (11, 20, "11-20"),
    (21, 50, "21-50"),
    (51, float('inf'), "50+")
]

dlc_dist_data = []
for min_dlc, max_dlc, label in dlc_ranges:
    count = ((df['dlc_count'] >= min_dlc) & (df['dlc_count'] <= max_dlc)).sum()
    pct = (count / len(df)) * 100
    print(f"  {label:15s}: {count:7,}개 ({pct:6.2f}%)")
    dlc_dist_data.append((label, count))

print(f"\n📊 DLC 최다 보유 게임 Top 10:")
top_dlc_games = df.nlargest(10, 'dlc_count')[['name', 'dlc_count', 'release_year']]
for i, (idx, row) in enumerate(top_dlc_games.iterrows(), 1):
    year = f"{int(row['release_year'])}" if pd.notna(row['release_year']) else "N/A"
    print(f"  {i:2d}. {row['name'][:45]:45s} - {row['dlc_count']:3d} DLCs ({year})")

# 그래프: DLC 개수 분포 (영어로)
fig, ax = plt.subplots(figsize=(10, 6))
labels, counts = zip(*dlc_dist_data)
colors = plt.cm.coolwarm(np.linspace(0.2, 0.8, len(labels)))
ax.bar(range(len(labels)), counts, color=colors, edgecolor='black', linewidth=0.5)
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=30, ha='right')
ax.set_ylabel('Number of Games', fontsize=11)
ax.set_title('DLC Count Distribution', fontsize=14, fontweight='bold')
ax.set_yscale('log')  # 로그 스케일
ax.grid(True, axis='y', alpha=0.3)
for i, v in enumerate(counts):
    ax.text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()

print("\n\n✅ SECTION 1 분석 완료!")
print("="*80)
```

```python
# =============================================================================
# SECTION 2: 💰 가격 & 언어 & 플랫폼
# =============================================================================
print("\n\n" + "="*80)
print("SECTION 2: Price & Language & Platform")
print("="*80)

# 2-1. 가격 분석
print("\n[2-1] Price (price_int) Detailed Analysis")
print("-" * 80)

print(f"\nPrice Data Status:")
free_count = (df['price_int'] == 0).sum()
no_info = (df['price_int'] == -1).sum()
error = (df['price_int'] == -2).sum()
paid = (df['price_int'] > 0).sum()

print(f"  Free (0):        {free_count:7,} ({free_count/len(df)*100:6.2f}%)")
print(f"  Paid (>0):       {paid:7,} ({paid/len(df)*100:6.2f}%)")
print(f"  No Info (-1):    {no_info:7,} ({no_info/len(df)*100:6.2f}%)")
print(f"  Error (-2):      {error:7,} ({error/len(df)*100:6.2f}%)")

paid_games = df[df['price_int'] > 0]['price_int']
if len(paid_games) > 0:
    print(f"\nPaid Game Price Statistics:")
    print(f"  Mean:      {paid_games.mean():>10,.0f} KRW")
    print(f"  Median:    {paid_games.median():>10,.0f} KRW")
    print(f"  Min:       {paid_games.min():>10,.0f} KRW")
    print(f"  Max:       {paid_games.max():>10,.0f} KRW")
    print(f"  Std Dev:   {paid_games.std():>10,.0f} KRW")
    
    print(f"\n  Percentiles:")
    for pct in [25, 50, 75, 90, 95, 99]:
        val = paid_games.quantile(pct/100)
        print(f"    {pct:2d}%: {val:>10,.0f} KRW")
    
    print(f"\nPrice Range Distribution (Paid Games: {len(paid_games):,}):")
    price_ranges = [
        (1, 1000, "Under 1K"),
        (1001, 5000, "1K-5K"),
        (5001, 10000, "5K-10K"),
        (10001, 15000, "10K-15K"),
        (15001, 20000, "15K-20K"),
        (20001, 30000, "20K-30K"),
        (30001, 50000, "30K-50K"),
        (50001, 100000, "50K-100K"),
        (100001, float('inf'), "Over 100K")
    ]
    
    price_dist_data = []
    for min_p, max_p, label in price_ranges:
        count = ((paid_games >= min_p) & (paid_games <= max_p)).sum()
        pct = (count / len(paid_games)) * 100
        print(f"  {label:15s}: {count:6,} ({pct:5.2f}%)")
        price_dist_data.append((label, count))

# 그래프: 가격 분포
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 가격대별 분포 (막대 그래프)
labels, counts = zip(*price_dist_data)
colors = plt.cm.YlOrRd(np.linspace(0.3, 0.9, len(labels)))
ax1.bar(range(len(labels)), counts, color=colors, edgecolor='black', linewidth=0.5)
ax1.set_xticks(range(len(labels)))
ax1.set_xticklabels(labels, rotation=45, ha='right')
ax1.set_ylabel('Number of Games', fontsize=11)
ax1.set_title('Price Range Distribution (Paid Games)', fontsize=14, fontweight='bold')
ax1.set_yscale('log')
ax1.grid(True, axis='y', alpha=0.3)
for i, v in enumerate(counts):
    ax1.text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=8)

# 오른쪽: 가격 히스토그램 (5만원 이하만)
paid_under_50k = paid_games[paid_games <= 50000]
ax2.hist(paid_under_50k, bins=50, color='steelblue', edgecolor='black', alpha=0.7)
ax2.axvline(paid_under_50k.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {paid_under_50k.mean():,.0f} KRW')
ax2.axvline(paid_under_50k.median(), color='green', linestyle='--', linewidth=2, label=f'Median: {paid_under_50k.median():,.0f} KRW')
ax2.set_xlabel('Price (KRW)', fontsize=11)
ax2.set_ylabel('Number of Games', fontsize=11)
ax2.set_title('Price Distribution (Under 50K KRW)', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 2-2. 플랫폼 지원
print("\n[2-2] Platform Support Analysis")
print("-" * 80)

platform_support = {'Windows': 0, 'Mac': 0, 'Linux': 0}
platform_combos = {
    'Windows Only': 0, 'Windows + Mac': 0, 'Windows + Linux': 0,
    'Mac + Linux': 0, 'All Three': 0,
    'Mac Only': 0, 'Linux Only': 0, 'No Info': 0
}

for idx, row in df.iterrows():
    if isinstance(row['platforms'], dict):
        win = row['platforms'].get('windows', False)
        mac = row['platforms'].get('mac', False)
        linux = row['platforms'].get('linux', False)
        
        if win: platform_support['Windows'] += 1
        if mac: platform_support['Mac'] += 1
        if linux: platform_support['Linux'] += 1
        
        if win and not mac and not linux:
            platform_combos['Windows Only'] += 1
        elif win and mac and not linux:
            platform_combos['Windows + Mac'] += 1
        elif win and not mac and linux:
            platform_combos['Windows + Linux'] += 1
        elif not win and mac and linux:
            platform_combos['Mac + Linux'] += 1
        elif win and mac and linux:
            platform_combos['All Three'] += 1
        elif not win and mac and not linux:
            platform_combos['Mac Only'] += 1
        elif not win and not mac and linux:
            platform_combos['Linux Only'] += 1
    else:
        platform_combos['No Info'] += 1

print(f"\nIndividual Platform Support:")
for platform, count in platform_support.items():
    pct = (count / len(df)) * 100
    print(f"  {platform:10s}: {count:7,} ({pct:6.2f}%)")

print(f"\nPlatform Combinations:")
for combo, count in platform_combos.items():
    pct = (count / len(df)) * 100
    print(f"  {combo:25s}: {count:7,} ({pct:6.2f}%)")

# 그래프: 플랫폼 지원 현황
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 개별 플랫폼 지원
platforms = list(platform_support.keys())
counts = list(platform_support.values())
colors = ['#0078D4', '#A3AAAE', '#F9AB00']
ax1.bar(platforms, counts, color=colors, edgecolor='black', linewidth=1)
ax1.set_ylabel('Number of Games', fontsize=11)
ax1.set_title('Platform Support (Individual)', fontsize=14, fontweight='bold')
ax1.grid(True, axis='y', alpha=0.3)
for i, v in enumerate(counts):
    pct = (v / len(df)) * 100
    ax1.text(i, v, f'{v:,}\n({pct:.1f}%)', ha='center', va='bottom', fontsize=10)

# 오른쪽: 플랫폼 조합 (상위 6개)
sorted_combos = sorted(platform_combos.items(), key=lambda x: x[1], reverse=True)[:6]
combo_labels, combo_counts = zip(*sorted_combos)
colors = plt.cm.Set2(range(len(combo_labels)))
ax2.barh(range(len(combo_labels)), combo_counts, color=colors)
ax2.set_yticks(range(len(combo_labels)))
ax2.set_yticklabels(combo_labels)
ax2.set_xlabel('Number of Games', fontsize=11)
ax2.set_title('Platform Combinations (Top 6)', fontsize=14, fontweight='bold')
ax2.invert_yaxis()
for i, v in enumerate(combo_counts):
    pct = (v / len(df)) * 100
    ax2.text(v, i, f' {v:,} ({pct:.1f}%)', va='center')

plt.tight_layout()
plt.show()

# 2-3. 언어 지원 분석
print("\n[2-3] Supported Languages Analysis")
print("-" * 80)

all_languages = []
for langs in df['supported_languages']:
    if isinstance(langs, list) and len(langs) > 0:
        all_languages.extend(langs)

lang_counts = Counter(all_languages)
print(f"\nTotal Unique Languages: {len(lang_counts)}")
print(f"Games with Language Info: {(df['supported_languages'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}")

print(f"\nTop 30 Supported Languages:")
for i, (lang, count) in enumerate(lang_counts.most_common(30), 1):
    pct = (count / len(df)) * 100
    print(f"  {i:2d}. {lang:30s}: {count:6,} ({pct:5.2f}%)")

# 그래프: Top 20 지원 언어
fig, ax = plt.subplots(figsize=(12, 8))
top_langs = lang_counts.most_common(20)
langs, counts = zip(*top_langs)
colors = plt.cm.tab20(range(len(langs)))
ax.barh(range(len(langs)), counts, color=colors)
ax.set_yticks(range(len(langs)))
ax.set_yticklabels(langs)
ax.set_xlabel('Number of Games', fontsize=11)
ax.set_title('Top 20 Supported Languages', fontsize=14, fontweight='bold')
ax.invert_yaxis()
for i, v in enumerate(counts):
    pct = (v / len(df)) * 100
    ax.text(v, i, f' {v:,} ({pct:.1f}%)', va='center', fontsize=9)
plt.tight_layout()
plt.show()

# 2-4. 음성 언어 분석
print("\n[2-4] Audio Languages Analysis")
print("-" * 80)

all_audio_langs = []
for langs in df['audio_languages']:
    if isinstance(langs, list) and len(langs) > 0:
        all_audio_langs.extend(langs)

audio_counts = Counter(all_audio_langs)
print(f"\nTotal Unique Audio Languages: {len(audio_counts)}")
print(f"Games with Audio Info: {(df['audio_languages'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}")

print(f"\nTop 20 Audio Languages:")
for i, (lang, count) in enumerate(audio_counts.most_common(20), 1):
    pct = (count / len(df)) * 100
    print(f"  {i:2d}. {lang:30s}: {count:6,} ({pct:5.2f}%)")

# 2-5. 한국어 지원 상세 분석
print("\n[2-5] Korean Language Support Detailed Analysis")
print("-" * 80)

kr_sub = (df['is_korean_supported'] == True).sum()
kr_dub = (df['is_korean_dubbed'] == True).sum()
kr_both = ((df['is_korean_supported'] == True) & (df['is_korean_dubbed'] == True)).sum()
kr_sub_only = ((df['is_korean_supported'] == True) & (df['is_korean_dubbed'] == False)).sum()
kr_none = ((df['is_korean_supported'] == False) & (df['is_korean_dubbed'] == False)).sum()

print(f"\nKorean Support Status:")
print(f"  Subtitles:       {kr_sub:7,} ({kr_sub/len(df)*100:6.2f}%)")
print(f"  Dubbing:         {kr_dub:7,} ({kr_dub/len(df)*100:6.2f}%)")
print(f"  Both:            {kr_both:7,} ({kr_both/len(df)*100:6.2f}%)")
print(f"  Subtitles Only:  {kr_sub_only:7,} ({kr_sub_only/len(df)*100:6.2f}%)")
print(f"  No Support:      {kr_none:7,} ({kr_none/len(df)*100:6.2f}%)")

# 그래프: 한국어 지원 현황
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 왼쪽: 파이 차트
kr_support_data = [kr_sub_only, kr_both, kr_none]
kr_labels = [f'Subtitles Only\n{kr_sub_only:,} ({kr_sub_only/len(df)*100:.1f}%)',
             f'Both Sub+Dub\n{kr_both:,} ({kr_both/len(df)*100:.1f}%)',
             f'No Support\n{kr_none:,} ({kr_none/len(df)*100:.1f}%)']
colors = ['#4CAF50', '#2196F3', '#9E9E9E']
ax1.pie(kr_support_data, labels=kr_labels, colors=colors, autopct='', startangle=90)
ax1.set_title('Korean Language Support Status', fontsize=14, fontweight='bold')

# 오른쪽: 막대 그래프
categories = ['Subtitles', 'Dubbing', 'Both', 'Sub Only', 'None']
values = [kr_sub, kr_dub, kr_both, kr_sub_only, kr_none]
colors = ['#4CAF50', '#2196F3', '#9C27B0', '#FF9800', '#9E9E9E']
ax2.bar(categories, values, color=colors, edgecolor='black', linewidth=1)
ax2.set_ylabel('Number of Games', fontsize=11)
ax2.set_title('Korean Support Breakdown', fontsize=14, fontweight='bold')
ax2.set_yscale('log')
ax2.grid(True, axis='y', alpha=0.3)
for i, v in enumerate(values):
    ax2.text(i, v, f'{v:,}', ha='center', va='bottom', fontsize=9)
ax2.tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()

print("\n\n✅ SECTION 2 분석 완료!")
print("="*80)
```

```python
# =============================================================================
# SECTION 3: 📝 설명 & 분류 (AI 임베딩용)
# =============================================================================
print("\n\n" + "="*80)
print("SECTION 3: Description & Classification (Tags/Genres/Categories)")
print("="*80)

# 3-1. 설명 텍스트 길이 분석
print("\n[3-1] Description Text Length Analysis")
print("-" * 80)

print(f"\nEnglish Description (short_description_en):")
print(f"  Mean:      {df['desc_en_length'].mean():.0f} chars")
print(f"  Median:    {df['desc_en_length'].median():.0f} chars")
print(f"  Min:       {df['desc_en_length'].min()} chars")
print(f"  Max:       {df['desc_en_length'].max()} chars")
print(f"  Std Dev:   {df['desc_en_length'].std():.0f} chars")

print(f"\n  Percentiles:")
for pct in [25, 50, 75, 90]:
    val = df['desc_en_length'].quantile(pct/100)
    print(f"    {pct:2d}%: {val:.0f} chars")

print(f"\nEnglish Description Length Distribution:")
length_ranges = [
    (0, 0, "No Description"),
    (1, 100, "Very Short (1-100)"),
    (101, 200, "Short (101-200)"),
    (201, 400, "Medium (201-400)"),
    (401, float('inf'), "Long (400+)")
]

en_length_data = []
for min_len, max_len, label in length_ranges:
    count = ((df['desc_en_length'] >= min_len) & (df['desc_en_length'] <= max_len)).sum()
    pct = (count / len(df)) * 100
    print(f"  {label:25s}: {count:7,} ({pct:6.2f}%)")
    en_length_data.append((label, count))

print(f"\nKorean Description (short_description_kr):")
print(f"  Mean:      {df['desc_kr_length'].mean():.0f} chars")
print(f"  Median:    {df['desc_kr_length'].median():.0f} chars")

kr_desc_valid = (df['desc_kr_length'] > 50).sum()
print(f"\n  Valid Korean Desc (>50 chars): {kr_desc_valid:,} ({kr_desc_valid/len(df)*100:.2f}%)")

# 3-2. 장르 (Genres) 분석
print("\n[3-2] Genres Analysis")
print("-" * 80)

all_genres = []
for genres in df['genres_en']:
    if isinstance(genres, list) and len(genres) > 0:
        all_genres.extend(genres)

genre_counts = Counter(all_genres)
print(f"\nTotal Unique Genres: {len(genre_counts)}")
print(f"Games with Genre Info: {(df['genres_en'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}")

print(f"\nTop 20 Popular Genres:")
for i, (genre, count) in enumerate(genre_counts.most_common(20), 1):
    pct = (count / len(df)) * 100
    print(f"  {i:2d}. {genre:30s}: {count:6,} ({pct:5.2f}%)")

# 장르 조합 분석
print(f"\nGenre Combinations (2-genre pairs):")
genre_combinations = []
for genres in df['genres_en']:
    if isinstance(genres, list) and len(genres) >= 2:
        sorted_genres = sorted(genres)
        for combo in combinations(sorted_genres, 2):
            genre_combinations.append(combo)

combo_counts = Counter(genre_combinations)
print(f"Total Unique Combinations: {len(combo_counts)}")
print(f"\nTop 15 Genre Combinations:")
for i, (combo, count) in enumerate(combo_counts.most_common(15), 1):
    print(f"  {i:2d}. {combo[0]:20s} + {combo[1]:20s}: {count:5,}")

# 그래프: 장르 분석
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# 왼쪽: Top 15 장르
top_genres = genre_counts.most_common(15)
genres, counts = zip(*top_genres)
colors = plt.cm.Spectral(np.linspace(0, 1, len(genres)))
bars1 = ax1.barh(range(len(genres)), counts, color=colors)
ax1.set_yticks(range(len(genres)))
ax1.set_yticklabels(genres)
ax1.set_xlabel('Number of Games', fontsize=11)
ax1.set_title('Top 15 Popular Genres', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, (bar, v) in enumerate(zip(bars1, counts)):
    pct = (v / len(df)) * 100
    ax1.text(bar.get_width(), bar.get_y() + bar.get_height()/2., 
             f' {v:,} ({pct:.1f}%)', va='center', fontsize=9)

# 오른쪽: Top 12 장르 조합
top_combos = combo_counts.most_common(12)
combo_labels = [f"{c[0][0][:10]}+{c[0][1][:10]}" for c in top_combos]
combo_values = [c[1] for c in top_combos]
colors = plt.cm.viridis(np.linspace(0.2, 0.9, len(combo_labels)))
bars2 = ax2.barh(range(len(combo_labels)), combo_values, color=colors)
ax2.set_yticks(range(len(combo_labels)))
ax2.set_yticklabels(combo_labels, fontsize=9)
ax2.set_xlabel('Number of Games', fontsize=11)
ax2.set_title('Top 12 Genre Combinations', fontsize=14, fontweight='bold')
ax2.invert_yaxis()
for i, (bar, v) in enumerate(zip(bars2, combo_values)):
    ax2.text(bar.get_width(), bar.get_y() + bar.get_height()/2., 
             f' {v:,}', va='center', fontsize=8)

plt.tight_layout()
plt.show()

# 3-3. 태그 (Tags) 분석 - 가장 중요!
print("\n[3-3] User Tags Analysis - Most Important for Recommendations!")
print("-" * 80)

all_tags = []
tag_positions = defaultdict(list)

for idx, tags in enumerate(df['tags_en']):
    if isinstance(tags, list) and len(tags) > 0:
        all_tags.extend(tags)
        for position, tag in enumerate(tags):
            tag_positions[tag].append(position)

tag_counts = Counter(all_tags)
print(f"\nTotal Unique Tags: {len(tag_counts)}")
print(f"Games with Tag Info: {(df['tags_en'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}")
print(f"Average Tags per Game: {len(all_tags) / len(df):.2f}")

print(f"\nTop 50 Popular Tags (with avg position):")
for i, (tag, count) in enumerate(tag_counts.most_common(50), 1):
    pct = (count / len(df)) * 100
    avg_pos = np.mean(tag_positions[tag]) if tag in tag_positions else 0
    print(f"  {i:2d}. {tag:30s}: {count:6,} ({pct:5.2f}%) | Avg Pos: {avg_pos:.1f}")

# 태그 조합 분석 (Top 5 태그만)
print(f"\nTag Combinations (Top 5 tags only):")
top_5_tags = [tag for tag, _ in tag_counts.most_common(5)]
print(f"Analyzing: {', '.join(top_5_tags)}")

tag_combinations = []
for tags in df['tags_en']:
    if isinstance(tags, list) and len(tags) >= 2:
        filtered_tags = [tag for tag in tags if tag in top_5_tags]
        if len(filtered_tags) >= 2:
            sorted_tags = sorted(filtered_tags)
            for combo in combinations(sorted_tags, 2):
                tag_combinations.append(combo)

tag_combo_counts = Counter(tag_combinations)
print(f"\nTop 10 Tag Combinations:")
for i, (combo, count) in enumerate(tag_combo_counts.most_common(10), 1):
    print(f"  {i:2d}. {combo[0]:20s} + {combo[1]:20s}: {count:5,}")

# 그래프: 태그 분석
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3)
ax1 = fig.add_subplot(gs[0, :])  # Top row - full width
ax2 = fig.add_subplot(gs[1, 0])  # Bottom left
ax3 = fig.add_subplot(gs[1, 1])  # Bottom right

# 상단: Top 30 태그
top_tags = tag_counts.most_common(30)
tags, counts = zip(*top_tags)
colors = plt.cm.rainbow(np.linspace(0, 1, len(tags)))
bars_top = ax1.barh(range(len(tags)), counts, color=colors)
ax1.set_yticks(range(len(tags)))
ax1.set_yticklabels(tags, fontsize=9)
ax1.set_xlabel('Number of Games', fontsize=11)
ax1.set_title('Top 30 User Tags', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, (bar, v) in enumerate(zip(bars_top, counts)):
    pct = (v / len(df)) * 100
    ax1.text(bar.get_width(), bar.get_y() + bar.get_height()/2., 
             f' {v:,} ({pct:.1f}%)', va='center', fontsize=7)

# 하단 왼쪽: 태그 평균 순위 (Top 20)
top_20_with_pos = [(tag, count, np.mean(tag_positions[tag])) for tag, count in tag_counts.most_common(20)]
tags_pos, counts_pos, avg_positions = zip(*top_20_with_pos)
avg_positions = np.array(avg_positions)
colors = plt.cm.coolwarm(avg_positions / np.max(avg_positions))
bars_pos = ax2.barh(range(len(tags_pos)), avg_positions, color=colors)
ax2.set_yticks(range(len(tags_pos)))
ax2.set_yticklabels(tags_pos, fontsize=8)
ax2.set_xlabel('Average Position (lower = more important)', fontsize=10)
ax2.set_title('Top 20 Tags - Average Position', fontsize=12, fontweight='bold')
ax2.invert_yaxis()
for i, (bar, v) in enumerate(zip(bars_pos, avg_positions)):
    ax2.text(bar.get_width(), bar.get_y() + bar.get_height()/2., 
             f' {v:.1f}', va='center', fontsize=7)

# 하단 오른쪽: Top 10 태그 조합
top_tag_combos = tag_combo_counts.most_common(10)
combo_labels = [f"{c[0][0][:12]}+\n{c[0][1][:12]}" for c in top_tag_combos]
combo_values = [c[1] for c in top_tag_combos]
colors = plt.cm.plasma(np.linspace(0.2, 0.9, len(combo_labels)))
bars_combo = ax3.bar(range(len(combo_labels)), combo_values, color=colors, edgecolor='black', linewidth=0.5)
ax3.set_xticks(range(len(combo_labels)))
ax3.set_xticklabels(combo_labels, rotation=45, ha='right', fontsize=7)
ax3.set_ylabel('Number of Games', fontsize=10)
ax3.set_title('Top 10 Tag Combinations', fontsize=12, fontweight='bold')
ax3.grid(True, axis='y', alpha=0.3)
for i, (bar, v) in enumerate(zip(bars_combo, combo_values)):
    ax3.text(bar.get_x() + bar.get_width()/2., bar.get_height(), 
             f'{v:,}', ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.show()

# 3-4. 카테고리 (Categories) 분석
print("\n[3-4] Categories Analysis")
print("-" * 80)

all_categories = []
for categories in df['categories_en']:
    if isinstance(categories, list) and len(categories) > 0:
        all_categories.extend(categories)

category_counts = Counter(all_categories)
print(f"\nTotal Unique Categories: {len(category_counts)}")
print(f"Games with Category Info: {(df['categories_en'].apply(lambda x: isinstance(x, list) and len(x) > 0)).sum():,}")

print(f"\nTop 20 Popular Categories:")
for i, (category, count) in enumerate(category_counts.most_common(20), 1):
    pct = (count / len(df)) * 100
    print(f"  {i:2d}. {category:40s}: {count:6,} ({pct:5.2f}%)")

# 멀티플레이어 관련 카테고리
print(f"\nMultiplayer-Related Categories:")
multiplayer_keywords = ['Multi-player', 'Co-op', 'Online', 'PvP', 'Multiplayer']
mp_categories = [cat for cat in category_counts.keys() 
                 if any(keyword in cat for keyword in multiplayer_keywords)]

for cat in sorted(mp_categories):
    count = category_counts[cat]
    pct = (count / len(df)) * 100
    print(f"  - {cat:40s}: {count:6,} ({pct:5.2f}%)")

# 그래프: 카테고리 분석
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# 왼쪽: Top 15 카테고리
top_cats = category_counts.most_common(15)
cats, counts = zip(*top_cats)
cats_short = [c[:35] + '...' if len(c) > 35 else c for c in cats]
colors = plt.cm.tab20b(np.linspace(0, 1, len(cats)))
bars_cat1 = ax1.barh(range(len(cats_short)), counts, color=colors)
ax1.set_yticks(range(len(cats_short)))
ax1.set_yticklabels(cats_short, fontsize=9)
ax1.set_xlabel('Number of Games', fontsize=11)
ax1.set_title('Top 15 Categories', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
for i, (bar, v) in enumerate(zip(bars_cat1, counts)):
    pct = (v / len(df)) * 100
    ax1.text(bar.get_width(), bar.get_y() + bar.get_height()/2., 
             f' {v:,} ({pct:.1f}%)', va='center', fontsize=8)

# 오른쪽: 멀티플레이어 카테고리
mp_cat_data = [(cat, category_counts[cat]) for cat in sorted(mp_categories)]
mp_cat_data.sort(key=lambda x: x[1], reverse=True)
mp_labels = [c[0][:25] + '...' if len(c[0]) > 25 else c[0] for c in mp_cat_data]
mp_values = [c[1] for c in mp_cat_data]
colors = plt.cm.Set3(np.linspace(0, 1, len(mp_labels)))
bars_cat2 = ax2.barh(range(len(mp_labels)), mp_values, color=colors)
ax2.set_yticks(range(len(mp_labels)))
ax2.set_yticklabels(mp_labels, fontsize=8)
ax2.set_xlabel('Number of Games', fontsize=11)
ax2.set_title('Multiplayer-Related Categories', fontsize=14, fontweight='bold')
ax2.invert_yaxis()
for i, (bar, v) in enumerate(zip(bars_cat2, mp_values)):
    pct = (v / len(df)) * 100
    ax2.text(bar.get_width(), bar.get_y() + bar.get_height()/2., 
             f' {v:,} ({pct:.1f}%)', va='center', fontsize=7)

plt.tight_layout()
plt.show()

# 3-5. 장르별 대표 태그 분석
print("\n[3-5] Representative Tags by Genre (Top 5 Genres)")
print("-" * 80)

top_5_genres = [genre for genre, _ in genre_counts.most_common(5)]

for genre in top_5_genres:
    print(f"\n[{genre}]")
    
    # 해당 장르를 가진 게임들의 태그 수집
    genre_tags = []
    genre_game_count = 0
    for idx, row in df.iterrows():
        if isinstance(row['genres_en'], list) and genre in row['genres_en']:
            genre_game_count += 1
            if isinstance(row['tags_en'], list):
                genre_tags.extend(row['tags_en'])
    
    genre_tag_counts = Counter(genre_tags)
    print(f"  Total games: {genre_game_count:,}")
    print(f"  Top 10 Tags:")
    for i, (tag, count) in enumerate(genre_tag_counts.most_common(10), 1):
        pct = (count / genre_game_count) * 100 if genre_game_count > 0 else 0
        print(f"    {i:2d}. {tag:25s}: {count:5,} ({pct:5.1f}%)")

print("\n\n✅ SECTION 3 분석 완료!")
print("="*80)
```

```python
# =============================================================================
# SECTION 4: 📊 외부 평가 지표 (메타크리틱 & 추천수)
# =============================================================================
print("\n\n" + "="*80)
print("SECTION 4: External Rating Metrics (Metacritic & Recommendations)")
print("="*80)

# 4-1. 메타크리틱 점수 분석
print("\n[4-1] Metacritic Score Analysis")
print("-" * 80)

metacritic_valid = df[df['metacritic'] > 0]
metacritic_missing = df[df['metacritic'] == 0]

print(f"\nMetacritic Data Availability:")
print(f"  With Score:    {len(metacritic_valid):7,} ({len(metacritic_valid)/len(df)*100:6.2f}%)")
print(f"  Missing:       {len(metacritic_missing):7,} ({len(metacritic_missing)/len(df)*100:6.2f}%)")

if len(metacritic_valid) > 0:
    print(f"\nMetacritic Score Statistics:")
    print(f"  Mean:      {metacritic_valid['metacritic'].mean():.1f}")
    print(f"  Median:    {metacritic_valid['metacritic'].median():.1f}")
    print(f"  Min:       {metacritic_valid['metacritic'].min()}")
    print(f"  Max:       {metacritic_valid['metacritic'].max()}")
    print(f"  Std Dev:   {metacritic_valid['metacritic'].std():.1f}")
    
    print(f"\n  Percentiles:")
    for pct in [25, 50, 75, 90]:
        val = metacritic_valid['metacritic'].quantile(pct/100)
        print(f"    {pct:2d}%: {val:.0f}")
    
    # 점수 등급별 분포
    print(f"\nScore Rating Distribution:")
    score_ranges = [
        (90, 100, "Masterpiece (90+)"),
        (80, 89, "Excellent (80-89)"),
        (70, 79, "Good (70-79)"),
        (60, 69, "Average (60-69)"),
        (0, 59, "Poor (<60)")
    ]
    
    score_dist_data = []
    for min_score, max_score, label in score_ranges:
        count = ((metacritic_valid['metacritic'] >= min_score) & 
                 (metacritic_valid['metacritic'] <= max_score)).sum()
        pct = (count / len(metacritic_valid)) * 100
        print(f"  {label:25s}: {count:5,} ({pct:5.2f}%)")
        score_dist_data.append((label, count))

# 그래프: 메타크리틱 점수 분포
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 점수 등급별 분포
labels, counts = zip(*score_dist_data)
colors = ['#4CAF50', '#8BC34A', '#FFC107', '#FF9800', '#F44336']
ax1.bar(range(len(labels)), counts, color=colors, edgecolor='black', linewidth=1)
ax1.set_xticks(range(len(labels)))
ax1.set_xticklabels(labels, rotation=30, ha='right')
ax1.set_ylabel('Number of Games', fontsize=11)
ax1.set_title('Metacritic Score Distribution by Rating', fontsize=14, fontweight='bold')
ax1.grid(True, axis='y', alpha=0.3)
for i, v in enumerate(counts):
    pct = (v / len(metacritic_valid)) * 100
    ax1.text(i, v, f'{v:,}\n({pct:.1f}%)', ha='center', va='bottom', fontsize=9)

# 오른쪽: 점수 히스토그램
ax2.hist(metacritic_valid['metacritic'], bins=30, color='steelblue', edgecolor='black', alpha=0.7)
ax2.axvline(metacritic_valid['metacritic'].mean(), color='red', linestyle='--', 
            linewidth=2, label=f"Mean: {metacritic_valid['metacritic'].mean():.1f}")
ax2.axvline(metacritic_valid['metacritic'].median(), color='green', linestyle='--', 
            linewidth=2, label=f"Median: {metacritic_valid['metacritic'].median():.1f}")
ax2.set_xlabel('Metacritic Score', fontsize=11)
ax2.set_ylabel('Number of Games', fontsize=11)
ax2.set_title('Metacritic Score Distribution (Histogram)', fontsize=14, fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 4-2. 추천수(Recommendations) 분석
print("\n[4-2] Steam Recommendations Analysis")
print("-" * 80)

recs_valid = df[df['recommendations_total'] > 0]
recs_zero = df[df['recommendations_total'] == 0]

print(f"\nRecommendations Data Availability:")
print(f"  With Recs:     {len(recs_valid):7,} ({len(recs_valid)/len(df)*100:6.2f}%)")
print(f"  Zero Recs:     {len(recs_zero):7,} ({len(recs_zero)/len(df)*100:6.2f}%)")

if len(recs_valid) > 0:
    print(f"\nRecommendations Statistics (games with recs):")
    print(f"  Mean:      {recs_valid['recommendations_total'].mean():>10,.0f}")
    print(f"  Median:    {recs_valid['recommendations_total'].median():>10,.0f}")
    print(f"  Min:       {recs_valid['recommendations_total'].min():>10,}")
    print(f"  Max:       {recs_valid['recommendations_total'].max():>10,}")
    
    print(f"\n  Percentiles:")
    for pct in [25, 50, 75, 90, 95, 99]:
        val = recs_valid['recommendations_total'].quantile(pct/100)
        print(f"    {pct:2d}%: {val:>10,.0f}")
    
    # 인기도 등급별 분포
    print(f"\nPopularity Rating Distribution:")
    popularity_ranges = [
        (100000, float('inf'), "Blockbuster (100K+)"),
        (50000, 99999, "Hit (50K-100K)"),
        (10000, 49999, "Very Popular (10K-50K)"),
        (5000, 9999, "Popular (5K-10K)"),
        (1000, 4999, "Moderate (1K-5K)"),
        (100, 999, "Normal (100-1K)"),
        (1, 99, "Niche (1-100)")
    ]
    
    pop_dist_data = []
    for min_rec, max_rec, label in popularity_ranges:
        count = ((recs_valid['recommendations_total'] >= min_rec) & 
                 (recs_valid['recommendations_total'] <= max_rec)).sum()
        pct = (count / len(recs_valid)) * 100
        print(f"  {label:30s}: {count:6,} ({pct:5.2f}%)")
        pop_dist_data.append((label, count))

# 그래프: 추천수 분포
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# 왼쪽: 인기도 등급별 분포
labels, counts = zip(*pop_dist_data)
colors = plt.cm.Reds(np.linspace(0.9, 0.3, len(labels)))
ax1.barh(range(len(labels)), counts, color=colors)
ax1.set_yticks(range(len(labels)))
ax1.set_yticklabels(labels, fontsize=9)
ax1.set_xlabel('Number of Games', fontsize=11)
ax1.set_title('Popularity Distribution', fontsize=14, fontweight='bold')
ax1.invert_yaxis()
ax1.set_xscale('log')
for i, v in enumerate(counts):
    pct = (v / len(recs_valid)) * 100
    ax1.text(v, i, f' {v:,} ({pct:.1f}%)', va='center', fontsize=8)

# 중간: 로그 스케일 히스토그램
recs_log = np.log10(recs_valid['recommendations_total'] + 1)
ax2.hist(recs_log, bins=50, color='coral', edgecolor='black', alpha=0.7)
ax2.set_xlabel('Log10(Recommendations + 1)', fontsize=11)
ax2.set_ylabel('Number of Games', fontsize=11)
ax2.set_title('Recommendations Distribution (Log Scale)', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)

# 오른쪽: 추천수 <10K만 (상세)
recs_under_10k = recs_valid[recs_valid['recommendations_total'] <= 10000]['recommendations_total']
ax3.hist(recs_under_10k, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
ax3.axvline(recs_under_10k.mean(), color='red', linestyle='--', 
            linewidth=2, label=f"Mean: {recs_under_10k.mean():.0f}")
ax3.axvline(recs_under_10k.median(), color='green', linestyle='--', 
            linewidth=2, label=f"Median: {recs_under_10k.median():.0f}")
ax3.set_xlabel('Recommendations', fontsize=11)
ax3.set_ylabel('Number of Games', fontsize=11)
ax3.set_title('Recommendations Distribution (<10K)', fontsize=14, fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 4-3. 메타크리틱 vs 추천수 상관관계
print("\n[4-3] Metacritic Score vs Recommendations Correlation")
print("-" * 80)

both_valid = df[(df['metacritic'] > 0) & (df['recommendations_total'] > 0)]

if len(both_valid) > 100:
    correlation = both_valid['metacritic'].corr(both_valid['recommendations_total'])
    print(f"\nPearson Correlation: {correlation:.3f}")
    
    if correlation > 0.3:
        print("  -> Positive correlation (higher score = more recommendations)")
    elif correlation < -0.3:
        print("  -> Negative correlation")
    else:
        print("  -> Weak correlation (score and popularity are independent)")
    
    # 점수대별 평균 추천수
    print(f"\nAverage Recommendations by Score Range:")
    score_bins = [0, 60, 70, 80, 90, 100]
    for i in range(len(score_bins)-1):
        min_s, max_s = score_bins[i], score_bins[i+1]
        subset = both_valid[(both_valid['metacritic'] >= min_s) & 
                           (both_valid['metacritic'] < max_s)]
        if len(subset) > 0:
            avg_recs = subset['recommendations_total'].mean()
            med_recs = subset['recommendations_total'].median()
            print(f"  {min_s:2d}-{max_s:2d}: Avg {avg_recs:>9,.0f} | Med {med_recs:>8,.0f} (n={len(subset):,})")

# 그래프: 메타크리틱 vs 추천수
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 산점도 (로그 스케일)
ax1.scatter(both_valid['metacritic'], np.log10(both_valid['recommendations_total'] + 1), 
            alpha=0.3, s=20, c=both_valid['metacritic'], cmap='RdYlGn')
ax1.set_xlabel('Metacritic Score', fontsize=11)
ax1.set_ylabel('Log10(Recommendations + 1)', fontsize=11)
ax1.set_title(f'Metacritic vs Recommendations (Correlation: {correlation:.3f})', 
              fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

# 오른쪽: 점수대별 평균 추천수 (박스플롯)
score_categories = []
rec_by_score = []
for i in range(len(score_bins)-1):
    min_s, max_s = score_bins[i], score_bins[i+1]
    subset = both_valid[(both_valid['metacritic'] >= min_s) & 
                       (both_valid['metacritic'] < max_s)]
    if len(subset) > 0:
        score_categories.append(f"{min_s}-{max_s}")
        rec_by_score.append(subset['recommendations_total'].values)

bp = ax2.boxplot(rec_by_score, labels=score_categories, patch_artist=True)
for patch, color in zip(bp['boxes'], plt.cm.RdYlGn(np.linspace(0.2, 0.8, len(bp['boxes'])))):
    patch.set_facecolor(color)
ax2.set_xlabel('Metacritic Score Range', fontsize=11)
ax2.set_ylabel('Recommendations', fontsize=11)
ax2.set_title('Recommendations Distribution by Score Range', fontsize=14, fontweight='bold')
ax2.set_yscale('log')
ax2.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 4-4. 가격대별 평점/인기도 분석
print("\n[4-4] Quality Metrics by Price Range")
print("-" * 80)

paid_games = df[df['price_int'] > 0].copy()

if len(paid_games) > 0:
    # 가격 구간 정의
    price_bins = [0, 5000, 10000, 20000, 50000, float('inf')]
    price_labels = ['0-5K', '5K-10K', '10K-20K', '20K-50K', '50K+']
    
    paid_games['price_range'] = pd.cut(paid_games['price_int'], 
                                        bins=price_bins, 
                                        labels=price_labels, 
                                        include_lowest=True)
    
    print(f"\nGames by Price Range:")
    for label in price_labels:
        count = (paid_games['price_range'] == label).sum()
        pct = (count / len(paid_games)) * 100
        print(f"  {label:15s}: {count:6,} ({pct:5.2f}%)")
    
    # 가격대별 메타크리틱 평균
    print(f"\nAverage Metacritic by Price Range (with score only):")
    for label in price_labels:
        subset = paid_games[(paid_games['price_range'] == label) & 
                           (paid_games['metacritic'] > 0)]
        if len(subset) > 0:
            avg_meta = subset['metacritic'].mean()
            print(f"  {label:15s}: Avg {avg_meta:5.1f} (n={len(subset):,})")
    
    # 가격대별 추천수 평균
    print(f"\nAverage Recommendations by Price Range (with recs only):")
    for label in price_labels:
        subset = paid_games[(paid_games['price_range'] == label) & 
                           (paid_games['recommendations_total'] > 0)]
        if len(subset) > 0:
            avg_recs = subset['recommendations_total'].mean()
            med_recs = subset['recommendations_total'].median()
            print(f"  {label:15s}: Avg {avg_recs:>9,.0f} | Med {med_recs:>8,.0f} (n={len(subset):,})")

# 그래프: 가격대별 품질 지표
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# 왼쪽: 가격대별 메타크리틱 평균
meta_by_price = []
price_labels_with_data = []
for label in price_labels:
    subset = paid_games[(paid_games['price_range'] == label) & 
                       (paid_games['metacritic'] > 0)]
    if len(subset) > 0:
        meta_by_price.append(subset['metacritic'].values)
        price_labels_with_data.append(label)

if meta_by_price:
    bp1 = ax1.boxplot(meta_by_price, labels=price_labels_with_data, patch_artist=True)
    for patch, color in zip(bp1['boxes'], plt.cm.YlOrRd(np.linspace(0.3, 0.9, len(bp1['boxes'])))):
        patch.set_facecolor(color)
    ax1.set_xlabel('Price Range (KRW)', fontsize=11)
    ax1.set_ylabel('Metacritic Score', fontsize=11)
    ax1.set_title('Metacritic Score by Price Range', fontsize=14, fontweight='bold')
    ax1.grid(True, axis='y', alpha=0.3)

# 오른쪽: 가격대별 추천수 평균
recs_by_price = []
for label in price_labels:
    subset = paid_games[(paid_games['price_range'] == label) & 
                       (paid_games['recommendations_total'] > 0)]
    if len(subset) > 0:
        recs_by_price.append(subset['recommendations_total'].values)

if recs_by_price:
    bp2 = ax2.boxplot(recs_by_price, labels=price_labels_with_data, patch_artist=True)
    for patch, color in zip(bp2['boxes'], plt.cm.Blues(np.linspace(0.3, 0.9, len(bp2['boxes'])))):
        patch.set_facecolor(color)
    ax2.set_xlabel('Price Range (KRW)', fontsize=11)
    ax2.set_ylabel('Recommendations', fontsize=11)
    ax2.set_title('Recommendations by Price Range', fontsize=14, fontweight='bold')
    ax2.set_yscale('log')
    ax2.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 4-5. 무료 vs 유료 게임 품질 비교
print("\n[4-5] Free vs Paid Games Quality Comparison")
print("-" * 80)

free_games = df[df['is_free'] == True]
paid_games_comp = df[df['price_int'] > 0]

print(f"\n[Free Games] (Total: {len(free_games):,})")

free_with_meta = free_games[free_games['metacritic'] > 0]
if len(free_with_meta) > 0:
    print(f"  Metacritic: Avg {free_with_meta['metacritic'].mean():.1f} (n={len(free_with_meta):,})")
else:
    print(f"  Metacritic: No data")

free_with_recs = free_games[free_games['recommendations_total'] > 0]
if len(free_with_recs) > 0:
    print(f"  Recommendations: Avg {free_with_recs['recommendations_total'].mean():,.0f}, Med {free_with_recs['recommendations_total'].median():,.0f}")

print(f"\n[Paid Games] (Total: {len(paid_games_comp):,})")

paid_with_meta = paid_games_comp[paid_games_comp['metacritic'] > 0]
if len(paid_with_meta) > 0:
    print(f"  Metacritic: Avg {paid_with_meta['metacritic'].mean():.1f} (n={len(paid_with_meta):,})")

paid_with_recs = paid_games_comp[paid_games_comp['recommendations_total'] > 0]
if len(paid_with_recs) > 0:
    print(f"  Recommendations: Avg {paid_with_recs['recommendations_total'].mean():,.0f}, Med {paid_with_recs['recommendations_total'].median():,.0f}")

# 그래프: 무료 vs 유료 비교
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# 왼쪽: 메타크리틱 비교
if len(free_with_meta) > 0 and len(paid_with_meta) > 0:
    data_meta = [free_with_meta['metacritic'].values, paid_with_meta['metacritic'].values]
    bp1 = ax1.boxplot(data_meta, labels=['Free', 'Paid'], patch_artist=True)
    bp1['boxes'][0].set_facecolor('#4CAF50')
    bp1['boxes'][1].set_facecolor('#2196F3')
    ax1.set_ylabel('Metacritic Score', fontsize=11)
    ax1.set_title('Metacritic: Free vs Paid Games', fontsize=14, fontweight='bold')
    ax1.grid(True, axis='y', alpha=0.3)

# 오른쪽: 추천수 비교 (로그 스케일)
if len(free_with_recs) > 0 and len(paid_with_recs) > 0:
    data_recs = [free_with_recs['recommendations_total'].values, paid_with_recs['recommendations_total'].values]
    bp2 = ax2.boxplot(data_recs, labels=['Free', 'Paid'], patch_artist=True)
    bp2['boxes'][0].set_facecolor('#4CAF50')
    bp2['boxes'][1].set_facecolor('#2196F3')
    ax2.set_ylabel('Recommendations', fontsize=11)
    ax2.set_title('Recommendations: Free vs Paid Games', fontsize=14, fontweight='bold')
    ax2.set_yscale('log')
    ax2.grid(True, axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# 4-6. Top 게임 목록
print("\n[4-6] Top Games by Recommendations and Metacritic")
print("-" * 80)

print(f"\nTop 20 Most Recommended Games:")
top_recs = df.nlargest(20, 'recommendations_total')[['name', 'recommendations_total', 'metacritic', 'price_int', 'is_free']]
print(f"{'Rank':4s} {'Game Name':50s} {'Recs':>12s} {'Meta':>6s} {'Price':>10s}")
print("-" * 90)
for i, (idx, row) in enumerate(top_recs.iterrows(), 1):
    name = row['name'][:48] if len(row['name']) > 48 else row['name']
    recs = f"{row['recommendations_total']:,}"
    meta = f"{row['metacritic']}" if row['metacritic'] > 0 else "N/A"
    
    if row['is_free']:
        price = "Free"
    elif row['price_int'] > 0:
        price = f"{row['price_int']:,} KRW"
    else:
        price = "N/A"
    
    print(f"{i:2d}.  {name:50s} {recs:>12s} {meta:>6s} {price:>10s}")

print(f"\n\nTop 20 Highest Rated Games (Metacritic):")
top_rated = df[df['metacritic'] > 0].nlargest(20, 'metacritic')[['name', 'metacritic', 'recommendations_total', 'price_int', 'is_free']]
print(f"{'Rank':4s} {'Game Name':50s} {'Meta':>6s} {'Recs':>12s} {'Price':>10s}")
print("-" * 90)
for i, (idx, row) in enumerate(top_rated.iterrows(), 1):
    name = row['name'][:48] if len(row['name']) > 48 else row['name']
    meta = f"{row['metacritic']}"
    recs = f"{row['recommendations_total']:,}" if row['recommendations_total'] > 0 else "N/A"
    
    if row['is_free']:
        price = "Free"
    elif row['price_int'] > 0:
        price = f"{row['price_int']:,} KRW"
    else:
        price = "N/A"
    
    print(f"{i:2d}.  {name:50s} {meta:>6s} {recs:>12s} {price:>10s}")

print("\n\n✅ SECTION 4 분석 완료!")
print("="*80)
```
